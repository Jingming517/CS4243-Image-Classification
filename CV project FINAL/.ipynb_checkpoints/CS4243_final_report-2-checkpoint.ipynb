# 1. Project Description

Our aim is to classify whether a particular image is computer generated imagery (CGI) or a photographic image (PI). CGI refers to still or animated content made with computer software and is typically used in movies, television and games. Photographic images refers to a "a picture made using a camera, in which an image is focused on to light-sensitive material and then made visible and permanent by chemical treatment, or stored digitally". 

In particular, we have decided to focus our project on images of human faces because we believe that this is where the greatest potential for abuse lies. In particular, recent advancements in computer graphics have allowed for realistic images of humans to be generated programmatically - this includes AI-generated imagery such as Deep Fakes. 

In an [experiment](https://www.scientificamerican.com/article/humans-find-ai-generated-faces-more-trustworthy-than-the-real-thing/) on over 200 participants, researchers found that the group could only differentiate AI generated images from real images 48.2% of the time.

Real life image of actor Peter Cushing on the left and CGI image of him on the right.

This is potentially dangerous because criminals can create images of people in incriminating situations (e.g. armed soldiers engaging in war crimes) and lead to civil distrust, undermining of public safety, manipulating elections etc. 

Thus, an effective way of discriminating CGI images from a real photographic image is needed to prevent abuses of such technology as it advances over time.


## Related works

Previously reported works have included the use of CNNs to discriminated between CGI and PI. 



Rahmouni et al. proposes the use of a novel method for this task that integrates a statistical feature extraction to a CNN frameworks and finds the best features for efficient boundary. The dataset was mainly trained on a photo-realistic video-game screenshots database. Using this model, the team claimed to have an accuracy of 84.8% when classifying images. It is worth noting that the work of Rahmouni et al. builds on the work of Wu et al., who proposed a statistical-distribution-based method to distinguish different types of images. The main intuition behind this idea is that CGI and PI can be distinguished by "utilizing some usual image statistics".  In their paper, the team computes histograms of the difference images (i.e. the images composed of differences of adjacent pixels) and afterwards takes the most informative histograms as features. Examples of these histograms are shown below. 




Screenshot 2022-04-25 at 1.20.08 PM.png

To further build on the work of Rahmouni et al, Nguyen et al. proposes the  use of a modular CGI-PI discriminator using a customized VGG-19 as the feature extractor, statistical convolutional neural networks as the feature transformers, and a discriminator. Using this approach, along with Fisher's linear discriminant analysis (LDA) the Nguyen et al. team claimed to have an accuracy of 88.95% when classifying images.

## Image resolution

Ideally, a high resolution image would definitely be ideal for training, however, the a 512 * 512 image size is 16 times larger than 128 * 128 image size. This was a huge scaling factor in terms of space we required.

In our case, we used 128 * 128 dimensions, and the training dataset was already 3 GB in size. Hence we decided not to go with larger dimensions.

A 256 * 256 dimension would have resulted in a 12 GB training dataset, and a 1024 * 1024 dimension would have resulted in a 196 GB training dataset.

Also, another problem we faced is that not all images were that high resolution, so it was better for us to stick with something that was manageable.

We opted for more variations in images for our training rather than higher resolution for training.

## Cropping vs resizing

Initially, we resized original images into a standard square format, but then we realised that it would not be good for training, as proportions will be distorted. The pixel definitions will also be affected.

Take for instance resizing an image of 1920 * 1024 to 128 * 128. Firstly, the objects and people in the image will shrink either vertically or horizontally, secondly, a lot of features in the higher resolution image will be lost when the size is shrunk. Hence, there exist problems that resizing will introduce, whereas centre cropping is a much preferred strategy.

We then finalised on the center cropping strategy: to crop from the center of the original image, a standard fixed size of 128 * 128 image. This way, the original image proportion is retained, and the definition is also retained.

This helps our model in training.

````python
# Change data path according to where your photos file is at.

# My directory: 
# base_folder
    # /photos
        # CGI - ...
        # CGI - ...
        # family photos
        # ...
    # /npy
    
# base_folder = os.path.join(os.path.join(os.path.expanduser('~')), 'data') # Change this to your base folder
base_folder = './data' # Change this to your base folder
photos_folder = base_folder + "/photos"
npy_folder = base_folder + "/npy"

image_matrix = [] # This will be your X, each entry is [red, green, blue, sobelx, sobely, sobelxy, edge_detect]
image_labels = [] # This will be your y labels
image_dimensions = 128 # Can change accordingly
total_num_of_features = 7

def count_total_images():
    total_size = 0
    for folder in os.listdir(photos_folder):
        if folder == '.DS_Store':
            continue
        for i, filename in enumerate(os.listdir(photos_folder + "/" + folder)):
            if filename.endswith(".jpeg") or filename.endswith(".png") or filename.endswith(".jpg"):
                total_size += 1
    return total_size


def load_image(infilename):
    """This function loads an image into memory when you give it
       the path of the image
    """
    img = cv2.imread(infilename)
#     show_nparray_as_img(img[...,::-1])
    center = [img.shape[0] / 2, img.shape[1] / 2]
    x = center[1] - image_dimensions / 2
    y = center[0] - image_dimensions / 2
    crop_img = img[int(y):int(y+image_dimensions), int(x):int(x+image_dimensions)]
#     resized_img = cv2.resize(img, (image_dimensions, image_dimensions), interpolation = cv2.INTER_AREA)
    crop_grey_img = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
    grey_img_blur = cv2.GaussianBlur(crop_grey_img, (3,3), 0)

    # # Sobel Edge Detection
    sobelx = cv2.Sobel(src=grey_img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
    sobely = cv2.Sobel(src=grey_img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
    sobelxy = cv2.Sobel(src=grey_img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
    # # Canny Edge Detection
    edges_detected = cv2.Canny(grey_img_blur , 100, 200) 
 
    blue_img = crop_img[:,:,0]
    green_img = crop_img[:,:,1]
    red_img = crop_img[:,:,2]

    return np.array([red_img, green_img, blue_img, sobelx, sobely, sobelxy, edges_detected])
    
def create_npy_from_image(photos_folder, output_file, total_size, image_dim):
    """Loops through the images in a folder and saves all of them
       as a numpy array in output_name
    """
    global image_matrix
    count = 0
    
    image_matrix = np.empty((total_size, total_num_of_features, image_dim, image_dim), dtype=np.float32)
    
    for folder in os.listdir(photos_folder):
        if folder == '.DS_Store':
            continue
        for i, filename in enumerate(os.listdir(photos_folder + "/" + folder)):
            if filename.endswith(".jpeg") or filename.endswith(".png") or filename.endswith(".jpg"):
                data = load_image(photos_folder + "/" + folder + "/" + filename)
                image_matrix[count] = data
                if "CGI" in folder:
                    image_labels.append(1)
                else:
                    image_labels.append(0)
                count += 1
            else:
                continue
    np.save(output_file, image_matrix)
````

### Alexnet

We tested the collected dataset on an existing model, Alexnet.  
The architecture consists of eight layers: five convolutional layers and three fully-connected layers.  

Since AlexNet has 60 million parameters, a major issue in terms of overfitting.  
AlexNet is considered as the milestone of CNN for image classification.
The convolution layers are used to extract the edges of the images and fully connected layers are used to learn these features.
However, AlexNet is not deep enough compared to the later models. The use of large convolution filters (5*5) is also not encouraged.

Since AlexNet had 60 million parameters, a major issue is overfitting.Two methods were introduced to reduce overfitting: Data Augmentation and Dropout. Dropout works by randomly setting the outgoing edges of hidden units (neurons that make up hidden layers) to 0 at each update of the training phase.

````python
# Model
# AlexNet
from typing import Any

import torch
import torch.nn as nn



__all__ = ["AlexNet", "alexnet"]


model_urls = {
    "alexnet": "https://download.pytorch.org/models/alexnet-owt-7be5be79.pth",
}


class AlexNet(nn.Module):
    def __init__(self, num_classes: int = 2, dropout: float = 0.2) -> None:
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=11, stride=4, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),
            nn.Conv2d(64, 192, kernel_size=5, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),
            nn.Conv2d(192, 384, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(384, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),
        )
        self.avgpool = nn.AdaptiveAvgPool2d((6, 6))
        self.classifier = nn.Sequential(
            nn.Dropout(p=dropout),
            nn.Linear(256 * 6 * 6, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(p=dropout),
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            nn.Linear(4096, num_classes),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.features(x)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x
````

### LeNet
The LeNet-5 architecture consists of two sets of convolutional and average pooling layers, followed by a flattening convolutional layer, then two fully-connected layers and finally a softmax classifier.

Features:  
Every convolutional layer includes three parts: convolution, pooling, and nonlinear activation functions  
Using convolution to extract spatial features  
Subsampling average pooling layer  
tanh activation function  
Using MLP as the last classifier  
Sparse connection between layers to reduce the complexity of computation

````python
# Model
# LeNet-5
class LeNet5(nn.Module):

    def __init__(self):

        super().__init__()
        # input: 3*128*128
        # block 1:         3 x 128 x 128 --> 32 x 32 x 32        
        self.conv1 = nn.Conv2d(3,   32,  kernel_size=5, padding=2 )
        self.pool1  = nn.AvgPool2d(4, stride=4)

        # block 2:         32 x 32 x 32  --> 64 x 8 x 8
        self.conv2 = nn.Conv2d(32,  64, kernel_size=5, padding=2 )
        self.pool2  = nn.AvgPool2d(4, stride=4)

        # linear layers:   64 x 8 x 8 --> 4096 --> 
        self.linear1 = nn.Linear(4096,2048)
        self.linear2 = nn.Linear(2048, 1024)
        self.linear3 = nn.Linear(1024, 2)

    def forward(self, x):

        # block 1:         
        x = self.conv1(x)
        x = torch.relu(x)
        x = self.pool1(x)

        # block 2:         
        x = self.conv2(x)
        x = torch.relu(x)
        x = self.pool2(x)


        x = x.reshape(-1, 4096)
        x = self.linear1(x)
        x = torch.relu(x)
        x = self.linear2(x)
        x = torch.relu(x)
        x = self.linear3(x) 
        
        return x
````

The drop-outp rates were selected through trial and error. This is to prevent overfitting problems.

````python
# Model: VGGNet
# CNN, can edit parameters
class pic_cartoon_net(nn.Module):

    def __init__(self):

        super().__init__()
        # input: 3*128*128
        # block 1:         3 x 128 x 128 --> 32 x 64 x 64        
        self.conv1 = nn.Conv2d(3,   32,  kernel_size=3, padding=1 )
        self.pool1  = nn.MaxPool2d(2,2)

        # block 2:         32 x 64 x 64 --> 64 x 32 x 32
        self.conv2 = nn.Conv2d(32,  64, kernel_size=3, padding=1 )
        self.pool2  = nn.MaxPool2d(2,2)
        self.drop2 = nn.Dropout(p=0.3) #regularization

        # block 3:         64 x 32 x 32 --> 128 x 16 x 16        
        self.conv3a = nn.Conv2d(64, 64, kernel_size=3, padding=1 )
        self.conv3b = nn.Conv2d(64, 128, kernel_size=5, padding=2 )
        self.pool3  = nn.MaxPool2d(2,2)
        self.drop3 = nn.Dropout(p=0.4)
        
        #block 4:          128 x 16 x 16  --> 128 x 8 x 8
        self.conv4 = nn.Conv2d(128, 128, kernel_size=5, padding=2 )
        self.pool4  = nn.MaxPool2d(2,2)
        self.drop4 = nn.Dropout(p=0.5)

        # linear layers:   128 x 8 x 8 --> 8192 --> 2048 --> 2048 --> 2
        self.linear1 = nn.Linear(8192, 4096)
        self.linear2 = nn.Linear(4096,2048)
        self.linear3 = nn.Linear(2048, 2)

    def forward(self, x):

        # block 1:         
        x = self.conv1(x)
        x = torch.relu(x)
        x = self.pool1(x)

        # block 2:         
        x = self.conv2(x)
        x = torch.relu(x)
        x = self.pool2(x)
        x = self.drop2(x)

        # block 3:         
        x = self.conv3a(x)
        x = torch.relu(x)
        x = self.conv3b(x)
        x = torch.relu(x)
        x = self.pool3(x)
        x = self.drop3(x)

        #block 4:          
        x = self.conv4(x)
        x = torch.relu(x)
        x = self.pool4(x)
        x = self.drop4(x)

        # linear layers:   128 x 8 x 8 --> 8192 --> 2048 --> 2048 --> 2
        x = x.reshape(-1, 8192)
        x = self.linear1(x)
        x = torch.relu(x)
        x = self.linear2(x)
        x = torch.relu(x)
        x = self.linear3(x) 
        
        return x
````

````python
def eval_on_test_set():

    running_error=0
    num_batches=0

    # To change the loop after knowing data size
    for i in range(0,1000,bs):

        minibatch_data =  X1_test[i:i+bs]
        minibatch_label= y1_test[i:i+bs]

#         minibatch_data=minibatch_data.to(device)
#         minibatch_label=minibatch_label.to(device)
        
        inputs = (minibatch_data - mean)/std

        scores=net( inputs ) 

        error = utils.get_error( scores , minibatch_label)

        running_error += error.item()

        num_batches+=1

    total_error = running_error/num_batches
    print( 'error rate on test set =', total_error*100 ,'percent')
    return total_error
````

````python
start=time.time()
epoch_hist = []
error_hist = []

for epoch in range(1,6):
    
    # divide the learning rate by 2 at epoch 10, 14 and 18
    if epoch==10 or epoch == 14 or epoch==18:
        my_lr = my_lr / 2
    
    # create a new optimizer at the beginning of each epoch: give the current learning rate.   
    optimizer=torch.optim.SGD( net.parameters() , lr=my_lr )
        
    # set the running quatities to zero at the beginning of the epoch
    running_loss=0
    running_error=0
    num_batches=0
    
    # set the order in which to visit the image from the training set
    shuffled_indices=torch.randperm(500)
 
    for count in range(0,500,bs):
    
        # Set the gradients to zeros
        optimizer.zero_grad()
        
        # create a minibatch       
        indices=shuffled_indices[count:count+bs]
        minibatch_data =  X1_train[indices]
        minibatch_label=  y1_train[indices]
        
        # send them to the gpu
#         minibatch_data=minibatch_data.to(device)
#         minibatch_label=minibatch_label.to(device)
        
        # normalize the minibatch (this is the only difference compared to before!)
        inputs = (minibatch_data - mean)/std

        # tell Pytorch to start tracking all operations that will be done on "inputs"
        inputs.requires_grad_()

        # forward the minibatch through the net 
        scores=net( inputs ) 

        # Compute the average of the losses of the data points in the minibatch
        loss =  criterion( scores , minibatch_label) 
        
        # backward pass to compute dL/dU, dL/dV and dL/dW   
        loss.backward()

        # do one step of stochastic gradient descent: U=U-lr(dL/dU), V=V-lr(dL/dU), ...
        optimizer.step()
        

        # START COMPUTING STATS
        
        # add the loss of this batch to the running loss
        running_loss += loss.detach().item()
        
        # compute the error made on this batch and add it to the running error       
        error = utils.get_error( scores.detach() , minibatch_label)
        running_error += error.item()
        
        num_batches+=1        
    
    
    # compute stats for the full training set
    total_loss = running_loss/num_batches
    total_error = running_error/num_batches
    elapsed = (time.time()-start)/60
    

    print('epoch=',epoch, '\t time=', elapsed,'min','\t lr=', my_lr  ,'\t loss=', total_loss , '\t error=', total_error*100 ,'percent')
    eval_on_test_set() 
    print(' ')
    epoch_hist.append(epoch)
    error_hist.append(eval_on_test_set())
````

## Explanation of model architecture
Our group’s final model outputs the lowest error of 15.9%.

Our model uses a similar but smaller architecture as VGGNet. We are not using a full VGG model because big models with a large number of parameters will cause overfitting problems and take longer time to train. 

Our model consists of 5 convolutional layers, 4 max pooling layers, 3 dropout layers and 3 linear layers. This is a relatively deeper network comparing to LeNet and AlexNet. Since our project is to classify between computer generated imagery (CGI) and photographic image (PI), we need to extract and identify the differences in complex details. Depth of neural networks may improve generalizability. Neural Networks with more layers are better are breaking down the connects it learns into more granular features, and reuse them for the concepts. The high-level features can be beneficial for our model’s performance. Besides, neural networks with more layers are able to model more relationships and have a larger receptive field.

### Logistic Regression
Logistic Regression does not perform well in this image classification problem. 
This model is too simple to extract useful features. Image features are very hierarchical and complex and one linear layer is not enough to extract enough features for the classification task. Besides, Linear Regression deals with continuous values whereas classification problems are discrete values.

### LeNet
LeNet does not do as well with color images. Most image recognition problems would require RGB images for better recognition. Since the model isn’t very deep, it struggles to scan for all features thus producing poor performing models. If the Neural Network isn’t fed with enough features from the training images then it would be difficult for the model to generalize and create an accurate model.

### AlexNet
Compared to LeNet, the AlexNet architecture are better at the following parts. Firstly, AlexNet uses ReLU activation function which accelerated learning signifcantly compared to tanh. Secondly, it applies max pooling instead of average pooling. Thirdly, it has overlapping pooling filters to reduce the size of the network which also resulted in a decrease in error. Lastly, the Dropout layers after the last pooling layer and the first fully connected layer to improve generalization and reduce overfitting.


## Limitations

### Size of dataset
One key limitation to this project is the size of the 
dataset. We had to scrape for data online as opposed to using datasets that were already available. This meant that we could not make use of well known datasets such as ImageNet, nor could we contact other authors to borrow/purchase their dataset. This meant that we had to create our own scraper and scrape our own dataset. The rate of scraping was slow to prevent being blocked by Google Images and we also had to spend extra time on de-duplicating and pre-processing to standardize this raw dataset. The result was a relatively small dataset of about 4500 photos. We believe that the model results will be greatly improved with a much larger dataset. Furthermore, if the dataset was greatly increased, we could have used the Vision Transformer architecture which works better on large datasets (see Future Work for more details).  

### Lack of standardized dataset
It is also difficult for us to benchmark ourselves against other papers due to the lack of a standardized dataset. Even though large image datasets such as ImageNet, CIFAR, Google's Open Images etc. exist, there isn't an especially large one for CGI vs PI. A step in the right direction would definitely be the collation of a large dataset by the community.


### Size of images 
Our image dimensions were 128 * 128. Ideally, we would have wanted to train the images in their original higher resolution, however due to the limitations on space, it would be very difficult to train on higher image resolutions. For instance, if we were to train on 1024 * 1024 images, the space required would be 3GB * 64 = 192GB, as compared to simply just 3GB for our current implementation.

We opted for more different types of images than a higher resolution image.

To alleviate the problem of low resolution, we used center cropping on images rather than resizing them. This is very important as mentioned in our Data exploration phase. Imagine resizing a 1920 * 1920 to 128 * 128 resolution, a lot of features would have been lost. To avoid this, we crop the image's center with our image dimensions instead. This way, the original image definition is maintained, and used for training.

# 7. Future work

While CNNs are currently the main models behind the highest performing solutions to this problem, they suffer from several drawbacks. The fact that the architecture of the CNNs are designed specifically for images and can be computationally expensive. Vision Transformers (ViT) promise a more domain agnostic and computationally efficient architectures to state-of-the-art results.

ViT is at model that is based on the Transformer achitecture (see figure below for the architecure) originally designed for text-based tasks (AKA natural language processing). 

the-transformer-block-vit.png

The ViT outlined by Google's research team works in the following steps: 

1.   Divide by an image into a grid of square patches (e.g. divide an image into 3 x 3 squares)
2.   Flatten the patches
3.   Produce lower-dimensional linear embeddings from the flattened patches
4.   Add positional embeddings
5.   Feed the sequence as an input to a standard transformer encoder (such as the one shown in the figure about)
6.   Pretrain the model with image labels
7.   Finetune on the downstream dataset for image classification



image2.png

Researchers first tried training the model on the smaller ImageNet dataset (1M images) and it performed poorly with only a score of 77.9% accuracy against state-of-the-art CNN with 85.8% accuracy. However, real results came when the model was trained on the ImageNet-21k dataset (14M images with 21k classes) and JFT dataset (300M images with 18k classes). It then began to outperform the previous models. This suggests that with enough datapoints, ViT has been proved to outperfrom the tradtional CNN model. 

Our group would certainly like to experiment with ViT and/or other models in the future and tweak it to be better than all previous models.
